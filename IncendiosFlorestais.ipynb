```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
Base_Dados=pd.read_csv('Dados_Indendio.csv', encoding='latin-1')
```

```python
Base_Dados.head()
```

```python
Base_Dados.isnull().sum()
```

```python
Base_Dados.nunique()
```

```python
Analise_01= Base_Dados.groupby(by=['year']).sum()
Analise_01.head(20)
```

```python
plt.figure(figsize=(12,5))
plt.style.use('ggplot')
plt.title('Incendios no Brasil do ano de 1997 até 2017', loc='left', fontsize=15)
sns.lineplot(data=Analise_01, x='year', y='number', estimator='sum', lw=2, color='#ff5555',alpha=0.85);
```

```python
Analise_02=Base_Dados.groupby(by=['state']).sum().reset_index()[['state','number']].sort_values('number', ascending=False)

Analise_02.head(10)

```

```python
plt.figure(figsize=(12,5))
plt.title('Analise de queimadas por estado', loc='left', fontsize=15)

plt.bar(Analise_02.state, Analise_02 ['number'], color='#ff4e3f')

#labels
plt.ylabel=("Quantidade")
plt.xticks(rotation=90);
```
